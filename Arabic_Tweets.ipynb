```
# تم تحديد الوظيفة لتنفيذ جميع تقنيات المعالجة المسبقة. 

# المعالجة المسبقة للبيانات هي تقنية تُستخدم لتحويل البيانات الأولية إلى مجموعة بيانات نظيفة.

def preprocessing(Arabic_text):

  # إزالة جميع الكلمات الإنجليزية

  Arabic_text = re.sub('[\0-\200]+', " ", Arabic_text)


  arabic_diacritics = re.compile("""
                             ّ    | # Tashdid
                             َ    | # Fatha
                             ً    | # Tanwin Fath
                             ُ    | # Damma
                             ٌ    | # Tanwin Damm
                             ِ    | # Kasra
                             ٍ    | # Tanwin Kasr
                             ْ    | # Sukun
                             ـ     # Tatwil/Kashida
                         """, re.VERBOSE)
  
  # تطبيع النص العربي

  Arabic_text = re.sub("[إأآا]", "ا", Arabic_text)
  Arabic_text = re.sub("ى", "ي", Arabic_text)
  Arabic_text = re.sub("ؤ", "ء", Arabic_text)
  Arabic_text = re.sub("ئ", "ء", Arabic_text)
  Arabic_text = re.sub("ة", "ه", Arabic_text)
  Arabic_text = re.sub("گ", "ك", Arabic_text)

  # إزالة علامات التشكيل النص

  Arabic_text = re.sub(arabic_diacritics, '', Arabic_text)

  # إزالة تكرار الحرف

  Arabic_text = re.sub(r'(.)\1+', r'\1', Arabic_text)

  # إزالة الرموز التعبيرية أو الملصقات من النص

  emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  
        u"\U0001F300-\U0001F5FF"  
        u"\U0001F680-\U0001F6FF"  
        u"\U0001F1E0-\U0001F1FF"  
                           "]+", flags=re.UNICODE)
  
  Arabic_text=emoji_pattern.sub(r'', Arabic_text) # no emoji

  return Arabic_text

Arabic_text=preprocessing(Arabic_text)

# الإخراج هنا نص عربي نظيف

print("الإخراج هنا نص عربي نظيف")

print()

print(Arabic_text)
```
